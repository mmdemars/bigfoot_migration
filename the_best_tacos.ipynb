```python
# Import our dependencies
import pandas as pd
```

```python
tacos_df = pd.read_csv("sources/better_tacos.csv")
tacos_df.head()
```

```python
tacos_df = tacos_df.dropna(subset = ['latitude', 'longitude'])
tacos_df.head()
```

```python
tacos_df = tacos_df.drop(["menupageurl", "pricerangemin", "pricerangemax", "id", "country"], axis=1, errors='ignore')
tacos_df.head()
```

```python
tacos_df.cuisines.unique()
```

```python
print(tacos_df.shape[0])
```

```python
tacos_df = tacos_df.drop(["Unnamed: 0", "menupageurl"], axis=1, errors='ignore')
tacos_df.head()
```

```python
tacos_df.name.unique()
```

```python
tacos_df.name.value_counts()
```

```python
chain_counts = tacos_df['name'].value_counts()
print(chain_counts)
```

```python
drop_chains = list(chain_counts[chain_counts > 5].index)

# Replace in dataframe
for place in drop_chains:
    tacos_df.name = tacos_df.name.replace(place,"dumb chain")
    
# Check to make sure drop was successful
tacos_df.name.value_counts()
```

```python
tacos_df.drop(tacos_df.index[tacos_df['name'] == "dumb chain"], inplace=True)
tacos_df.head()
```

```python
print(tacos_df.shape[0])
```

```python
#export shiny clean csv
from pathlib import Path  
filepath = Path('sources/best_tacos.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
tacos_df.to_csv(filepath, index=True)
```
