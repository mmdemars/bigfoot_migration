```python
#  Import and read the bfro-report-locations.csv.
import pandas as pd 
bigfoot_df = pd.read_csv("sources/bigfoot_sightings.csv")
bigfoot_df.head()
```

```python
import datetime

bigfoot_df['timestamp'] = pd.to_datetime(bigfoot_df['timestamp'])
bigfoot_df.head()
```

```python
bigfoot_df = bigfoot_df.drop(['Unnamed: 0'], axis=1)
bigfoot_df.head()
```

```python
#sort rows by time
migration_df = bigfoot_df.sort_values(by = 'timestamp')
migration_df.head()
```

```python
migration_df['month'] = migration_df["timestamp"].dt.month
migration_df.head()
```

```python
migration_df['year'] = migration_df["timestamp"].dt.year
migration_df.head()
```

```python
# sightings in each month
migration_df.month.value_counts()
```

```python
# sightings in each year
migration_df.year.value_counts()
```

```python
# classifications
migration_df.classification.value_counts()
```

```python
print(migration_df.loc[migration_df["year"] == 2053])
```
