```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Define the features set.
X = skeleton_df.copy()
X = X.drop("month", axis=1)
X.head()
```

```python
# Define the target set.
y = skeleton_df["month"].values
y[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Determine the shape of our training and testing sets.
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Splitting into Train and Test sets into an 80/20 split.
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y, random_state=78, train_size=0.80)
```

```python
# Determine the shape of our training and testing sets.
print(X_train2.shape)
print(X_test2.shape)
print(y_train2.shape)
print(y_test2.shape)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Creating the decision tree classifier instance.
model = tree.DecisionTreeClassifier()
# Fitting the model.
model = model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = model.predict(X_test_scaled)
```

```python
predictions
```

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)
cm
```

```python
# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1", "Actual 2", "Actual 3", "Actual 4", "Actual 5","Actual 6", "Actual 7", "Actual 8","Actual 9", "Actual 10", "Actual 11"], columns=["Predicted 0", "Predicted 1", "Predicted 2", "Predicted 3", "Predicted 4", "Predicted 5", "Predicted 6", "Predicted 7", "Predicted 8", "Predicted 9", "Predicted 10", "Predicted 11"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Define the features set.
X = skeleton_df.copy()
X = X.drop("month", axis=1)
X.head()
```

```python
# Define the target set.
y = skeleton_df["month"].values
y[:5]
```

```python
# Splitting into Train and Test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=500, random_state=78) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test_scaled)
```

```python
predictions
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1", "Actual 2", "Actual 3", "Actual 4", "Actual 5","Actual 6", "Actual 7", "Actual 8","Actual 9", "Actual 10", "Actual 11"], columns=["Predicted 0", "Predicted 1", "Predicted 2", "Predicted 3", "Predicted 4", "Predicted 5", "Predicted 6", "Predicted 7", "Predicted 8", "Predicted 9", "Predicted 10", "Predicted 11"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```
