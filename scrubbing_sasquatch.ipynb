```python
#  Import and read the bfro-report-locations.csv.
import pandas as pd 
bigfoot_df = pd.read_csv("sources/bfro-report-locations.csv")
bigfoot_df.head()
```

```python
#new dataframe for fixing the report text
title_blowout = pd.DataFrame(bigfoot_df["title"])
title_blowout.head()
```

```python
# create new column for the report numner
title_blowout['report_no'] = title_blowout['title'].str.extract(r'^(.*?):')
title_blowout.head()
```

```python
# remove Report number from title column
#title_blowout['title'] = title_blowout['title'].str.strip(r'^(.*?):')
#title_blowout['title'] = title_blowout['title'].map(lambda x: x.lstrip(r'^(.*?):').rstrip(':'))
title_blowout['title'] = title_blowout['title'].str.split(r':').str.get(1)

title_blowout.head()
```

```python
#remove Report from report-no column
title_blowout['report_no'] = title_blowout['report_no'].str.split(r'Report ').str.get(1)

title_blowout.head()
```

```python
# Import stopwords with nltk.
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
#remove stop words from report
title_blowout['key_words'] = title_blowout['title'].apply(lambda x: ' '
                                                         .join([word for word in x.split() if word not in (stop)]))
title_blowout.head()
```

```python
bigfoot_df['report'] = title_blowout['title']
bigfoot_df.head()
```

```python
bigfoot_df['key_words'] = title_blowout['key_words']
bigfoot_df.head()
```

```python
#drop columns
bigfoot_df = bigfoot_df.drop(['title'], axis=1)
bigfoot_df.head()
```

```python
import datetime

bigfoot_df['timestamp'] = pd.to_datetime(bigfoot_df['timestamp'])
bigfoot_df.head()
```

```python
#export shiny clean csv
from pathlib import Path  
filepath = Path('sources/bigfoot_sightings.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
bigfoot_df.to_csv(filepath, index=True)
```
